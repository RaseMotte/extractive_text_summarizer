```python
import bs4 as BeautifulSoup
import os
import re
import pprint
```

```python
path = "Summary-P2FE/html/"
dmoz_desc_path = "resume/dmoz-fr-description"

descriptions = {}

files = [(path + f) for f in os.listdir(path)]
for pathfile in files:
    print(pathfile)
    with open(pathfile, "r", encoding="utf8") as fd :
        doc = []
        for l in fd:
            doc.append(l)
            process = l.strip() == "</html>"
            if process :
                fulltext = " ".join(doc)
                soup = BeautifulSoup.BeautifulSoup(fulltext)
                s = soup.find("link", {"rel" : "canonical"})
                if s is not None:
                    regex = re.search("href=\"https?://([^\"]+)\"", str(s))
                    if regex:
                        url = regex.group(1)
                        found = False
                        with open(dmoz_desc_path, "r", encoding="utf8") as desc_file:
                            for l in desc_file:
                                if url in l:
                                    found = True
                                    descriptions[url] = l
                                    break
                            if not found:
                                meta_desc = soup.find("meta", {"name" : "description"})
                                meta_desc = meta_desc if meta_desc is not None else soup.find("meta", {"property" : "og:description"})
                                try:
                                    meta_desc = '' if meta_desc is None else meta_desc["content"]
                                except KeyError:
                                    meta_desc = '' if meta_desc is None else meta_desc["value"]
                                descriptions[url] = meta_desc
                doc = []
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(descriptions)
```
